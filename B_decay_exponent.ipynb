```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
import scipy.signal as scs
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 



""" B=0 """


a = np.loadtxt("B=0\\powerdata_1e+18K_0.txt", delimiter=",")

no_files = 1000

total_power = np.zeros([no_files,len(a)])

for ix in range(no_files):
  total_power[ix,:] = np.loadtxt(f"powerdata_1e+18K_{ix}.txt", delimiter=",")

power = sum(total_power)/no_files



plt.figure(2, figsize=(24,7))
plt.subplot(1,2,1)
plt.loglog(power, "c") # label=np.round(T[tt],3))

ind = len(power)

pl = lambda x, amp, index: amp * (x**index)

half_ind = int(np.floor(.5*ind))

x = np.linspace(1,len(power),len(power))[:half_ind]
#y = (np.abs(power)**2)[:half_ind]
y = power[:half_ind]
co = "bgrcmykbgrcmykbgrcmyk"

#plt.figure(1, figsize=(16,9))
#plt.loglog(x,y, co[tt],  label=np.round(T[tt],3))

logx = np.log10(x[70:-50])
logy = np.log10(y[70:-50])
yerr = 0.2 * y; 
logyerr = yerr / y

fitfunc = lambda p, x1: p[0] + p[1] * x1   
errfunc = lambda p, x2, y2: (y2 - fitfunc(p, x2)) 

pinit = [1.0, -1.0]
out = sc.optimize.leastsq(errfunc, pinit, args=(logx, logy), full_output=1)

pfinal = out[0]
covar = out[1]
# print(pfinal)
#print(covar)

index_1 = pfinal[1]
amp = 10.0**pfinal[0]

#indexErr = np.sqrt(covar[0][0])
#ampErr = np.sqrt(covar[1][1])*amp

print(f"the gradient for B=0 is {index_1}")
#plt.figure()
#plt.loglog(x[3:-20], pl(x, amp, index)[3:-20], co[tt]+"--", linewidth="2.5", label="Fitted curve {}".format(np.round(T[tt],3)))
#plt.figure(2, figsize=(16,9))
#plt.subplot(1,2,1)
plt.title('power spectra for full frequency range', fontsize="18")
plt.xlabel("frequency", fontsize="18")
plt.ylabel('noise spectra', fontsize="18")
plt.loglog(x[3:-20], pl(x, amp, index_1)[3:-20], co[0]+"--", linewidth="2.5")
#len(x)
plt.loglog(x[70],y[70], "kx", markersize=30)
plt.loglog(x[-50], y[-50], "kx", markersize=30)

#plt.figure(3, figsize=(16,9))
#plt.legend(loc="best")#
plt.subplot(1,2,2)
plt.title('power spectra up to nyquist frequency', fontsize="18")
plt.xlabel("frequency", fontsize="18")
plt.ylabel('noise spectra', fontsize="18")
plt.loglog(power[:half_ind], "c")
plt.loglog(x[9:-20], pl(x, amp, index_1)[9:-20], co[0]+"--", linewidth="2.5")
plt.loglog(x[70],y[70], "kx", markersize=30)
plt.loglog(x[-50], y[-50], "kx", markersize=30)
#plt.legend(loc="best")#
```
